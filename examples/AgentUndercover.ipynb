# AgentUndercover backend example notebook

### **1. step:** import *AgentUndercover* class

```python
from matheUI.backend import AgentUndercover
```

### **2. step:** create *AgentUndercover* class object
The two parameters **places_folder** and **places_files** will be automatically filled with the variables in **matheUI/config/\_\_init\_\_.py**

You can add more datasets with adding the pickled dictonary in **places_folder** and add the file name in **places_files** in **matheUI/config/\_\_init\_\_.py**

```python
auc = AgentUndercover()
```

### **3. step:** get data report 
(place groups, places and professions)

whole data report

```python
auc.data_report()
```

get methods for groups of places

```python
auc.get_places_groups()
```

get method for places of one group

```python
auc.get_places("standard")
```

get method for professions of one place

```python
auc.get_professions("standard", "Stadtreinigung")
```

### **4. step:** change data in class object **(changes not data in data files)**

add new place

```python
auc.add_place(places_group="standard", place_name="Fantasiewelt", professions=["Drache", "Zwerg", "Einhorn", "Zyklop", "Hexe"])
```

```python
auc.get_places("standard")
```

add new professions to a place

```python
auc.add_profession(places_group="standard", place_name="Fantasiewelt", profession="Magier")
```

```python
auc.get_professions("standard", "Fantasiewelt")
```

delete profession

```python
auc.delete_profession("standard", "Fantasiewelt", "Hexe")
```

```python
auc.get_professions("standard", "Fantasiewelt")
```

update place (if you want to change multiple professions for a place, use **add_place** and overwrite existing place)

```python
auc.add_place("standard", "Fantasiewelt", ['Feuerdrache', 'Zwerg', 'Einhorn', 'Zyklop', 'Elementarmagier'])
```

```python
auc.get_professions("standard", "Fantasiewelt")
```

delete place

```python
auc.delete_place("standard", "Fantasiewelt")
```

```python
auc.get_places("standard")
```

### **5. step:** update data files

You can update the data files with the changes you made in the class object. Before doing so, look through the data report and check if the data is fine.

```python
auc.update_data_files()
```

### **6. step:** generate round

**places_groups** default is **"all"**, change it to use your prefered places groups (list of places groups)

**Note:** professions can always be repetitive to give the spy a chance to mislead

```python
auc.generate_round(player_number=7, spy_number=2, places_groups="all")
```

```python
auc.generate_round(player_number=7, spy_number=2, places_groups=["standard"])
```
