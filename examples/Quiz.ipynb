# Quiz backend example notebook

### **1. step:** import *Quiz* class

```python
from matheUI_backend import Quiz
```

### **2. step:** create *Quiz* class object
The two parameters **data_folder_path** and **data_files** will be automatically filled with the variables in **matheUI/config/\_\_init\_\_.py**

You can add more datasets with adding the csv files in **data_folder_path** and add the file name in **data_files** in **matheUI/config/\_\_init\_\_.py**

```python
quiz = Quiz()
```

### **3. step:** get data report

whole data

```python
quiz.data_report()
```

get categories

```python
quiz.get_categories()
```

get question dataframe

```python
quiz.get_df("history events")
```

### **4. step:** change data in *Quiz* class object **(does not change the data files)**

create questions dataframe (static function)

```python
q_df = Quiz.create_questions_df(questions=["erstes iPhone"], solutions=["2007"], fake_answers1=["2005"], fake_answers2=["2009"], fake_answers3=["2011"], infos=["offizielle Ankündigung 09.01."])
q_df
```

add questions dataframe to category

```python
quiz.add_questions(category="history events", question_df=q_df)
```

```python
quiz.get_df("history events")
```

delete question

```python
quiz.delete_question(category="history events", row_index=37)
```

```python
quiz.get_df("history events")
```

### **5. step:** update data files

You can update the data files with the changes you made in the class object. Before doing so, look through the data report and check if the data is fine.

```python
quiz.update_data_files()
```

### **6. step:** generate round

**categories** default is **"all"**, change it to use your prefered categories (list of categories or name of category)

```python
quiz.generate_round(question_number=10, categories="all")
```

```python
quiz.generate_round(question_number=10, categories="history events")
```
